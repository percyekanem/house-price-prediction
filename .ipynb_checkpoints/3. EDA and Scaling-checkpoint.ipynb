# Data Pre-Processing

> First, the dataset is split into x and y components.


```python
x = hp_index_final.drop("AvgPrice", axis=1)
y= hp_index_final["AvgPrice"]
```

## Scaling

> Non-categorical data is standardized using the standard scaler, which uses the z-score to scale the data, giving it a mean of 0 and standard deviation of 1. (or normalized using the minmax scaler, which gives a range between 0 and 1 using the minumum and maximum values in the dataset. Check for outliers and normality first).
https://towardsdatascience.com/one-hot-encoding-standardization-pca-data-preparation-steps-for-segmentation-in-python-24d07671cf0b

> Normalization is usually used when data does not follow a normal distribution or the algorithm to be used does not assume a normal distribution (e.g. K-nearest neighbours and neural networks). But this method usually scales outliers down to a very small range.

> Standardization is usually used when the data is normal or the algorithm assumes normality. But the range of values among features varies in this method.

> The dependent variable will not be scaled, as this will disrupt the interpretability of the model results.

```python
features_to_scale = ["LaggedIndex","LaggedMonthChange","LaggedYearChange"]
```

```python
# Standardize

standardizer = StandardScaler()
x_stand = x.copy()
x_stand[features_to_scale] = standardizer.fit_transform(x_stand[features_to_scale])
x_stand[features_to_scale].describe()
```

Test both scaling methods with a model to determine the best one

```python
# Normalize

normalizer = MinMaxScaler()
x_norm = x.copy()
x_norm[features_to_scale] = normalizer.fit_transform(x_norm[features_to_scale])
x_norm[features_to_scale].describe()
```

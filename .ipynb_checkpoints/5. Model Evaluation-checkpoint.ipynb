```python
evaluate[["RMSE","NRMSE","R2"]] = evaluate[["RMSE","NRMSE","R2"]].apply(lambda x: round(x,4))
evaluate["Time"] = round(evaluate["Time"], 2)
evaluate["model_scaling"] = evaluate["Model"] + "_" + evaluate["Scaling"]
```

```python
evaluate
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="model_scaling", y="Time", data=evaluate.sort_values("Time"))
plt.xticks(rotation=90)
plt.xlabel("Model")
plt.ylabel("Time")
plt.title("Time Taken for Execution (seconds)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Model", y="NRMSE", data=evaluate.sort_values("NRMSE"), hue="Scaling")

plt.xlabel("Model")
plt.ylabel("NRMSE")
plt.title("Model Performance (NRMSE)")
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python
eval_standardised = evaluate[(evaluate["Scaling"]=="Standardised")&(~evaluate["Model"].str.contains("LR"))]

plt.figure(figsize=(10, 6))
sns.barplot(x="Model", y="NRMSE", data=eval_standardised.sort_values("NRMSE"), hue="Scaling")

plt.xlabel("Model")
plt.ylabel("NRMSE")
plt.title("Deep Learning Model Performance (NRMSE)")
plt.xticks(rotation=90)
plt.ylim(0, 0.26)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Model", y="R2", data=evaluate.sort_values("R2"), hue="Scaling")
plt.xticks(rotation=90)
plt.xlabel("Model")
plt.ylabel("R-squared score")
plt.title("Model Performance (R-squared)")
plt.ylim(0.6, 1)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Model", y="R2", data=evaluate.sort_values("R2"), hue="Scaling")
plt.xticks(rotation=90)
plt.xlabel("Model")
plt.ylabel("R-squared score")
plt.title("Model Performance (R-squared)")
plt.ylim(0.99, 1)
plt.tight_layout()
plt.show()
```
